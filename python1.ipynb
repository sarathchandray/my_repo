```python
import numpy as np

data = [3.4, -9.2, 2.4, -1.4, 5.5]

mean = np.mean(data)
std = np.std(data)
min = np.min(data)
max = np.max(data)
median = np.median(data)

print(f"mean : {mean}\nstd : {std}\nmin : {min}\nmax : {max}\nmedian : {median}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

xdata = []
ydata = []

for x in range(1,11) :
    xdata.append(x)
    ydata.append(np.sin(x))

plt.plot(xdata, ydata)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('sin')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

start = 0
end = np.pi*20
print(f"np.pi : {np.pi}")
increment = 0.001
xdata = np.arange(start, end, increment)
ydata = np.sin(xdata)

plt.plot(xdata, ydata)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('sin')
plt.grid()
plt.show()
```
